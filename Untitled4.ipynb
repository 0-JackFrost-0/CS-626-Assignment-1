```python
import nltk
from nltk.corpus import brown
from nltk import FreqDist
from nltk.tag import hmm
import numpy as np
import pandas as pd
from sklearn.model_selection import KFold
from sklearn.metrics import confusion_matrix, classification_report, precision_recall_fscore_support, fbeta_score
import matplotlib.pyplot as plt
import seaborn as sns
from collections import defaultdict

# Download Brown corpus and universal tagset if not already downloaded
nltk.download('brown')
nltk.download('universal_tagset')

# Load the Brown corpus with the "universal" tagset
corpus = brown.tagged_sents(tagset='universal')

def preprocess_data(corpus):
    processed_corpus = []
    for sent in corpus:
        processed_sentence = [(word.lower(), tag) for word, tag in sent]
        processed_corpus.append(processed_sentence)
    return processed_corpus

corpus = preprocess_data(corpus)

# Define all unique POS tags in the dataset
tags = list(set(tag for sent in corpus for _, tag in sent))

# Define training and test sets using 5-fold cross-validation
kf = KFold(n_splits=5, shuffle=True, random_state=42)
```

```python
class HMM_POS_Tagger:
    def __init__(self, corpus, tags):
        self.corpus = corpus
        self.tags = tags
        self.tag2idx = {tag: i for i, tag in enumerate(tags)}
        self.idx2tag = {i: tag for i, tag in enumerate(tags)}

        # Initialize transition, emission, and start probabilities
        self.transition_probs = defaultdict(lambda: defaultdict(lambda: 1e-5))
        self.emission_probs = defaultdict(lambda: defaultdict(lambda: 1e-5))
        self.start_probs = defaultdict(lambda: 1e-5)
        self.word_counts = defaultdict(int)
        self.tag_counts = defaultdict(int)

    def train(self, train_data):
        for sent in train_data:
            # Handle start of sentence
            self.start_probs[sent[0][1]] += 1
            prev_tag = None

            for word, tag in sent:
                self.word_counts[word] += 1
                self.tag_counts[tag] += 1

                # Handle emission probabilities
                self.emission_probs[tag][word] += 1

                # Handle transition probabilities
                if prev_tag:
                    self.transition_probs[prev_tag][tag] += 1
                prev_tag = tag

        # Normalize start probabilities
        total_starts = sum(self.start_probs.values())
        for tag in self.start_probs:
            self.start_probs[tag] /= total_starts

        # Normalize transition probabilities
        for tag1 in self.transition_probs:
            total = sum(self.transition_probs[tag1].values())
            for tag2 in self.transition_probs[tag1]:
                self.transition_probs[tag1][tag2] /= total

        # Normalize emission probabilities
        for tag in self.emission_probs:
            total = sum(self.emission_probs[tag].values())
            for word in self.emission_probs[tag]:
                self.emission_probs[tag][word] /= total

    def viterbi(self, sentence):
        # Viterbi algorithm for POS tagging
        V = [{}]
        path = {}

        # Initialize base cases (t == 0)
        for tag in self.tags:
            V[0][tag] = self.start_probs[tag] * self.emission_probs[tag].get(sentence[0], 1e-5)
            path[tag] = [tag]

        # Run Viterbi for t > 0
        for t in range(1, len(sentence)):
            V.append({})
            new_path = {}

            for curr_tag in self.tags:
                (prob, state) = max(
                    (V[t-1][prev_tag] * self.transition_probs[prev_tag].get(curr_tag, 1e-5) * self.emission_probs[curr_tag].get(sentence[t], 1e-5), prev_tag)
                    for prev_tag in self.tags
                )
                V[t][curr_tag] = prob
                new_path[curr_tag] = path[state] + [curr_tag]

            path = new_path

        # Find the final best path
        n = len(sentence) - 1
        (prob, state) = max((V[n][tag], tag) for tag in self.tags)
        return path[state]

    def tag_sentence(self, sentence):
        return self.viterbi(sentence)
```

```python
# Initialize the HMM POS Tagger
hmm_tagger = HMM_POS_Tagger(corpus, tags)

# Perform 5-Fold Cross-Validation
accuracies = []
confusion_matrices = []
precision_scores = []
recall_scores = []
f1_scores = []
f05_scores = []
f2_scores = []
fold = 1

for train_idx, test_idx in kf.split(corpus):
    train_data = [corpus[i] for i in train_idx]
    test_data = [corpus[i] for i in test_idx]

    hmm_tagger.train(train_data)

    y_true = []
    y_pred = []

    for sent in test_data:
        words, true_tags = zip(*sent)
        pred_tags = hmm_tagger.tag_sentence(words)
        y_true.extend(true_tags)
        y_pred.extend(pred_tags)

    # Calculate accuracy
    accuracy = np.mean(np.array(y_true) == np.array(y_pred))
    accuracies.append(accuracy)

    # Generate confusion matrix
    cm = confusion_matrix(y_true, y_pred, labels=tags)
    confusion_matrices.append(cm)

    # Compute precision, recall, and F1 score
    precision, recall, f1, _ = precision_recall_fscore_support(y_true, y_pred, labels=tags, average='weighted')
    f05 = fbeta_score(y_true, y_pred, beta=0.5, labels=tags, average='weighted')
    f2 = fbeta_score(y_true, y_pred, beta=2, labels=tags, average='weighted')

    # Per
    # TODO
    pos_metrics = precision_recall_fscore_support(y_true, y_pred, labels=tags)

    for i, tag in enumerate(tags):
      print(f"{tag}: Precision = {pos_metrics[0][i]:.4f}, Recall = {pos_metrics[1][i]:.4f}, F1 = {pos_metrics[2][i]:.4f}")

    precision_scores.append(precision)
    recall_scores.append(recall)
    f1_scores.append(f1)
    f05_scores.append(f05)
    f2_scores.append(f2)

    print(f"Fold {fold} Accuracy: {accuracy:.4f}")
    fold += 1
```

```python
# Calculate overall accuracy
overall_accuracy = np.mean(accuracies)
print(f"Overall Accuracy: {overall_accuracy:.4f}")

# Calculate confusion matrix and per POS accuracy
confusion_matrix_sum = np.sum(confusion_matrices, axis=0)
per_pos_accuracy = confusion_matrix_sum.diagonal() / confusion_matrix_sum.sum(axis=1)

# Calculate overall precision, recall, and F-scores
overall_precision = np.mean(precision_scores)
overall_recall = np.mean(recall_scores)
overall_f1 = np.mean(f1_scores)
overall_f05 = np.mean(f05_scores)
overall_f2 = np.mean(f2_scores)

print(f"Overall Precision: {overall_precision:.4f}")
print(f"Overall Recall: {overall_recall:.4f}")
print(f"Overall F1-score: {overall_f1:.4f}")
print(f"Overall F0.5-score: {overall_f05:.4f}")
print(f"Overall F2-score: {overall_f2:.4f}")

# Plot Confusion Matrix
plt.figure(figsize=(10, 8))
sns.heatmap(confusion_matrix_sum, annot=True, fmt='d', xticklabels=tags, yticklabels=tags, cmap='Blues')
plt.title('Confusion Matrix for POS Tagging')
plt.xlabel('Predicted')
plt.ylabel('True')
plt.show()

# Report per POS accuracy and performance metrics
# TODO : average and print per pos metrics
pos_metrics = precision_recall_fscore_support(y_true, y_pred, labels=tags)
print("\nPer POS Tag Performance (Precision, Recall, F1-Score):")
for i, tag in enumerate(tags):
    print(f"{tag}: Precision = {pos_metrics[0][i]:.4f}, Recall = {pos_metrics[1][i]:.4f}, F1 = {pos_metrics[2][i]:.4f}")
```
